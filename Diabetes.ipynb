```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X[['Pregnancies', 'Age']], y, test_size=0.30, random_state=0, stratify=y)
print ('Treningowe obserwacje: %d\nTestowe obserwacje: %d' % (X_train.shape[0], X_test.shape[0]))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

scaler_2var = StandardScaler()
X_train_standardized = scaler_2var.fit_transform(X_train)
X_test_standardized = scaler_2var.transform(X_test)
model_lr = LogisticRegression()
model_lr.fit(X_train_standardized, y_train)
```

```python
from numpy import where
from numpy import meshgrid
from numpy import arange
from numpy import hstack

def plot_classification_surface(X_plot, y_plot, trained_model):
    plt.figure(figsize=(12, 7))

    # okreÅ›lenie granic zbioru
    min1, max1 = X_plot[:, 0].min()-1, X_plot[:, 0].max()+1
    min2, max2 = X_plot[:, 1].min()-1, X_plot[:, 1].max()+1

    # skalowanie dla obu osi
    x1grid = arange(min1, max1, 0.1)
    x2grid = arange(min2, max2, 0.1)

    # utworzenie siatki
    xx, yy = meshgrid(x1grid, x2grid)

    # przetworzenie siatki w wektor
    r1, r2 = xx.flatten(), yy.flatten()
    r1, r2 = r1.reshape((len(r1), 1)), r2.reshape((len(r2), 1))

    # stworzenie zmiennych objaÅ›niajÄ…cych dla modelu
    grid = hstack((r1,r2))

    # predykcja modelu zbioru
    yhat = trained_model.predict(grid)

    # przetworzenie predykcji na decyzji dla konkretnego punktu wykresu
    zz = yhat.reshape(xx.shape)

    # wizualizacja siatki z przyporzÄ…dkowanÄ… klasÄ…
    plt.contourf(xx, yy, zz, cmap='Paired')

    # tworzenie wykresu punktowego dla klas ze zbioru X
    for class_value in range(2):

        # przyporzÄ…dkowanie klasy (y) do obsewacji (X)
        row_ix = where(y_plot == class_value)

        # stworzenie wykresu punktowego
        plt.scatter(X_plot[row_ix, 0], X_plot[row_ix, 1], cmap='Paired', alpha=0.3, label=class_value)

    # wizualizacja wykresu
    plt.legend(loc='upper right')
    plt.show()

plot_classification_surface(X_plot=X_train_standardized, y_plot=y_train, trained_model=model_lr)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, f1_score, roc_auc_score

# Tworzenie pustego DataFrame na metryki
metrics_dataframe = pd.DataFrame(columns=['Model', 'F1_score', 'AUC'])
models = []
models_names = []
predictions_proba_list = []

def calculate_metrics(model, name, X_checked, y_checked):
    global metrics_dataframe
    
    # Dodajemy model i jego nazwÄ™ do listy
    models.append(model)
    models_names.append(name)
    
    # Generowanie predykcji
    predictions = model.predict(X_checked)
    predictions_proba = model.predict_proba(X_checked)
    predictions_proba_list.append(predictions_proba[:,1])

    # WyÅ›wietlanie raportu metryk (Precision, Recall, F1, Accuracy)
    print(classification_report(y_checked, predictions))

    # Macierz konfuzji
    plt.figure()
    cm = confusion_matrix(y_checked, predictions)
    ax = sns.heatmap(cm, annot=True, cmap='Blues', fmt='.0f')
    ax.set_title('Confusion Matrix\n\n')
    ax.set_xlabel('\nPredicted Values')
    ax.set_ylabel('Actual Values')
    plt.show()

    # Rysowanie krzywej ROC
    fig = plt.figure(figsize=(6, 6))
    plt.plot([0, 1], [0, 1], 'k--')
    for model_selected, name_selected, pred_proba in zip(models, models_names, predictions_proba_list):
        fpr, tpr, thresholds = roc_curve(y_checked, pred_proba)
        plt.plot(fpr, tpr, label=name_selected)
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')
    plt.legend(loc='lower right')
    plt.show()

    # Obliczanie F1 i AUC
    f1_metric = f1_score(y_checked, predictions)
    auc_metric = roc_auc_score(y_checked, predictions_proba[:,1])
    
    # Tworzenie nowego wiersza jako DataFrame
    new_data = pd.DataFrame({
        'Model': [name],
        'F1_score': [f1_metric],
        'AUC': [auc_metric]
    })
    
    # UsuniÄ™cie ewentualnych pustych wartoÅ›ci z new_data
    new_data = new_data.dropna(how='all', axis=1)
    
    # Dodawanie nowych danych do metrics_dataframe
    metrics_dataframe = pd.concat([metrics_dataframe, new_data], ignore_index=True)
    
    return metrics_dataframe
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import f1_score
import matplotlib.pyplot as plt
import numpy as np

# Defining features and target variable
features = ['Pregnancies', 'PlasmaGlucose', 'DiastolicBloodPressure', 'TricepsThickness', 'SerumInsulin', 'BMI', 'DiabetesPedigree', 'Age']
target = 'Diabetic'
X, y = diabetes[features], diabetes[target]

# Splitting the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X[['Pregnancies', 'Age']], y, test_size=0.30, random_state=0, stratify=y)

# Standardizing the data
scaler = StandardScaler()
X_train_standardized = scaler.fit_transform(X_train)
X_test_standardized = scaler.transform(X_test)

# Function to train model with different regularizations and plot surface
def train_and_plot(X_train, y_train, X_test, y_test, penalty, C_value):
    # Initialize model with the given penalty and C
    model_lr = LogisticRegression(penalty=penalty, C=C_value, solver='saga', max_iter=1000)
    model_lr.fit(X_train, y_train)

    # Plot classification surface
    plot_classification_surface(X_train, y_train, model_lr)

    # Predict on training and test sets
    predictions_train = model_lr.predict(X_train)
    predictions_test = model_lr.predict(X_test)

    # Calculate F1 score for both training and test sets
    f1_train = f1_score(y_train, predictions_train)
    f1_test = f1_score(y_test, predictions_test)
    
    print(f"Penalty: {penalty}, C: {C_value}")
    print(f"F1 score (train): {f1_train}")
    print(f"F1 score (test): {f1_test}")
    print("-" * 50)

# Helper function for plotting classification surface
def plot_classification_surface(X_plot, y_plot, trained_model):
    plt.figure(figsize=(12, 7))

    # Define the plot boundaries
    min1, max1 = X_plot[:, 0].min() - 1, X_plot[:, 0].max() + 1
    min2, max2 = X_plot[:, 1].min() - 1, X_plot[:, 1].max() + 1

    # Generate grid for plotting
    x1grid = np.arange(min1, max1, 0.1)
    x2grid = np.arange(min2, max2, 0.1)
    xx, yy = np.meshgrid(x1grid, x2grid)

    # Prepare the grid data for model prediction
    r1, r2 = xx.flatten(), yy.flatten()
    grid = np.hstack((r1.reshape(-1, 1), r2.reshape(-1, 1)))

    # Predict using the trained model
    yhat = trained_model.predict(grid)

    # Reshape predictions back to the grid shape
    zz = yhat.reshape(xx.shape)

    # Plot the surface
    plt.contourf(xx, yy, zz, cmap='Paired')

    # Plot actual data points (fixing the warning and indentation)
    for class_value in range(2):
        row_ix = np.where(y_plot == class_value)
        plt.scatter(X_plot[row_ix, 0], X_plot[row_ix, 1], alpha=0.6, label=f"Class {class_value}")

    plt.legend(loc='upper right')
    plt.show()

# List of penalties and regularization strengths to test
penalties = ['l1', 'l2']
C_values = [0.01, 0.1, 1, 10]

# Testing different regularization methods and strengths
for penalty in penalties:
    for C_value in C_values:
        train_and_plot(X_train_standardized, y_train, X_test_standardized, y_test, penalty, C_value)
```

Observations:
1. As C increases (less regularization), both the train and test F1 scores increase, but only slightly.
2. The results between L1 and L2 regularization are quite similar.
3. There is no significant overfitting observed, as F1 scores on the train and test sets are relatively close.
